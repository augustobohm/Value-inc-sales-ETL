### Working the calculations

```python
#Cost per transaction column calculation
```

```python
CostPerItem	= df['CostPerItem']
```

```python
NumberOfItemsPurchased	= df['NumberOfItemsPurchased']
```

```python
SellingPricePerItem = df['SellingPricePerItem']
```

```python
CostPerTransaction = CostPerItem*NumberOfItemsPurchased
print(CostPerTransaction)
```

```python
SellingPricePerTransaction = SellingPricePerItem * NumberOfItemsPurchased
```

```python
df['CostPerTransaction'] = CostPerTransaction
df['SellingPricePerTransaction'] = SellingPricePerTransaction

df.info
```

```python
df['SalesPerTransaction'] = df['SellingPricePerItem'] * df['NumberOfItemsPurchased']
```

```python
df['ProfitPerTransaction'] = df['SalesPerTransaction'] - df['CostPerTransaction']
print(df['ProfitPerTransaction'])
```

```python
#Markup = (sales - cost)/cost
```

```python
df['Markup'] = df['ProfitPerTransaction'] / df['CostPerTransaction']
print(df['Markup'])
```

```python
df.head()
```

```python
df['Markup'] = round(df['Markup'],2)
df.head()
```

```python
#Adjusting the date of the dataframe
```

```python
df.dtypes
```

```python
df['Day'] = df['Day'].astype(str)
df['Year'] = df['Year'].astype(str)
df['Date'] = df['Day'] +"-" + df['Month'] +"-"+ df['Year']
print(df['Date'])
```

```python
#Splitting Client Keywords column
```

```python
split_col = df['ClientKeywords'].str.split(',', expand=True)
split_col.head()
```

```python
df['ClientAge'] = split_col[0]
df['ClientType'] = split_col[1]
df['LengthofContract'] = split_col[2]
df.head()
```

```python
#replace the brackets of the new columns
```

```python
df['ClientAge'] = df['ClientAge'].str.replace('[','')
df['LengthofContract'] = df['LengthofContract'].str.replace(']','')
df.head()
```

```python
#Adding seasons file
```

```python
seasons = pd.read_csv ('value_inc_seasons.csv', sep=';')
seasons.head()
```

```python
#merging the files
```

```python
merge_df = pd.merge(df, seasons,on='Month')
merge_df.head()
```

```python
merge_df=merge_df.drop('ClientKeywords', axis=1)
merge_df=merge_df.drop('Day', axis=1)
merge_df.head()
```

```python
merge_df.to_csv('ValueInc_Cleaned.csv', index=False)
```
